### Clean Up Script

This script takes all files on my desktop without a folder and sorts them into images, music, info, text and other. It uses a CNN to sort screenshots into music vs. info and regular expressions and simple Python to do the rest. 

**Requirements**
1. Have the image classifier `desktop_1.pkl` in a folder on your desktop  
2. have this notebook in that same folder


**Imports**

```python
import os
from fastai import *
from fastai.vision import *
import re 
```

**Main Method**

```python
not_folder = re.compile('.\.\w+') #regular expression that defines a file vs. a folder

for filename in os.listdir(path): 
    if(re.search(not_folder, filename)): #skips over folders
        sort(filename)
```

**Functions**

```python
#takes a file and sorts it accordingly
def sort(file):

    if(file.startswith('Screen Shot')):
        
        #load CNN
        learn = load_learner(script_folder + "/", model_name)
        
        #load image
        if(file.endswith('.png')): img = open_image(file)
        else: img = open_image(file + ".png")
       
        img = img.resize(224)
    
        #get prediction
        pred_class,pred_idx,outputs = learn.predict(img)
        
        classes = ['info', 'music'] #might need to switch these around
        folder = classes[np.argmax(outputs)]
        
        move(file,folder)
    
    else:
        
        folder = locate_folder(file)
        move(file,folder)
        

#determine which folder it goes into
def locate_folder(file):
    if(file.endswith('.pdf') or file.endswith('.docx') or file.endswith('doc')):
        return 'text'
    
    if(file.endswith('.png') or file.endswith('.jpg') or file.endswith('.jpeg') 
       or file.endswith('.png') or file.endswith('.JPG')):
        return 'images'
    
    return 'other'

#Moves the file into the corresponding folder
def move(file,folder):
    
    folder = f + "/" + folder
    current = path + "/" + file
    dest = path + "/" + folder + "/" + file
    os.rename(current, dest)
```
