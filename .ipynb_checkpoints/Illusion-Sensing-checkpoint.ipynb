What I want to do is determine whether an image recognition model can learn the 'laws of physics' that determine whether something is physically possible. I feebily attempted to do this through training the model on illusion and non-illusion drawings.

Although the model achieved a low error rate, this is likely due to the fact that the two classes of data are relatively distinct with one having straighter lines and more white space and the other comprised of a range of images. Also, it is likely that there were a lot of duplicates of images.

I could improve it by removing such duplicates and gathering cleaner training data.

```python
classes = ['illusion','non-illusion']
```

*Next, place the* **csv files** *into the* **directory**

**Downloading the Images**

*Note: If it doesn't work the first time, add max_workers = 0 to debug*

```python

path = Path('data/real')
folder = 'illusion'
file = 'illusion.csv'
dest = path/folder
download_images(path/file, dest, max_pics=200,max_workers=0)

path = Path('data/real')
folder = 'non-illusion'
file = 'non-illusion.csv'
dest = path/folder
download_images(path/file, dest, max_pics=200)
```

**Cleaning up the images**

```python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

**Loading the images as data**

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=2).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

**Train Model**

```python
model = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
model.fit_one_cycle(4)
```

```python
model.save('model-1')
```

*The model appears to have a learning rate that is too low as the training loss is greater than the validatation loss , indicating that we aren't fitting the data enough*

```python
model.unfreeze()
```

```python
model.lr_find()
```

```python
model.recorder.plot()
```

```python
model.fit_one_cycle(4, max_lr=slice(3e-5,5e-3))
```

```python
model.save('model-4')
```

Our model with the lowest error_rate appears to be model 2. <br>
**Let's load this model and see what the results indicate.**

```python
model.load('model-2');
```

```python
interp = ClassificationInterpretation.from_learner(model)
```

```python
interp.plot_confusion_matrix()
```

*It appears that our model predicts people celebrities as non-celebrities more often than the predicting non-celebrities as celebrities.*
*Let's visualize some of these errors*

```python
interp.plot_top_losses(k=2)
```

*Interesting, so it's not perfect, because the one on the right is clearly not an illusion*
